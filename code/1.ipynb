```python
from moviepy.editor import VideoFileClip

def convert_avi_to_mp4(input_file, output_file):
    # Загружаем видеофайл .avi
    clip = VideoFileClip(input_file)

    # Сохраняем в формате .mp4
    clip.write_videofile(output_file, codec='libx264')

# Пример использования функции
input_path = "/home/dmitrii/Desktop/hockey_akbars/code/runs/detect/predict/акбарс_цска_короткое.avi"   # Замените на путь к вашему .avi файлу
output_path = "/home/dmitrii/Desktop/hockey_akbars/code/runs/detect/predict/акбарс_цска_короткое.mp4"  # Замените на путь для сохранения .mp4 файла

convert_avi_to_mp4(input_path, output_path)
```
