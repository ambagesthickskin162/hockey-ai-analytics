```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20, 14))


image = cv2.cvtColor(cv2.imread('image.png'), cv2.COLOR_BGR2RGB)


image_coordinates = np.float32([
    [0.2955, 0.384],
    [0.623, 0.44],
    [0.59, 0.67],

    # [0.41, 0.2791],
])

player_coordinates = np.float32([[[0.58, 0.55]]])

real_coordinates = np.float32([
    [0.824, 0.276],
    [0.920, 0.5],
    [0.824, 0.724],

    # [0.934, 0.03],
])

h, w, _ = image.shape

for (x, y) in image_coordinates:
    image = cv2.circle(image, center=(int(x*w),int(y*h)), radius=10, color=(255, 0, 0), thickness=-1)

image = cv2.circle(image, center=(int(player_coordinates[0][0][0]*w),int(player_coordinates[0][0][1]*h)), radius=10, color=(0, 255, 0), thickness=-1)

plt.imshow(image)
plt.show()
```

```python
import cv2
import numpy as np

# Рассчитываем матрицу аффинного преобразования (используем только 3 точки)
M_affine = cv2.getAffineTransform(image_coordinates, real_coordinates)

# Преобразуем координаты игрока с помощью аффинного преобразования
player_real_coords_affine = cv2.transform(player_coordinates, M_affine)

# Результат преобразования
x_player_real_affine, y_player_real_affine = player_real_coords_affine[0][0]

print(f"Аффинное преобразование: Координаты игрока в реальной системе координат: ({x_player_real_affine}, {y_player_real_affine})")
```
